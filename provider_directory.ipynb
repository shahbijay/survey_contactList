```python
import pandas as pd
import requests
```

```python
df = pd.read_csv("DACNationalProvider.csv")
df
```

```python
df.columns
```

```python
# Drop columns not needed and rename columns
pro_new = df[['NPI', 'lst_nm', 'frst_nm', 'Cred', 'Med_sch', 'adr_ln_1', 'adr_ln_2', 'cty', 'st', 'zip', 'phn_numbr',
              'adrs_id']].rename(columns={'lst_nm': 'Last_Name', 'frst_nm': 'First_Name', 'Cred': 'Type_of_Licensure',
                                          'adr_ln_1': 'Address_1', 'adr_ln_2': 'Address_2', 'cty': 'City', 'st': 'State',
                                          'zip': 'Zip', 'phn_numbr': 'Phone', 'adrs_id': 'PseudoAddressID'})
```

```python
pro_new
```

```python
# Select providers from Kentucky only.
providers_ky = pro_new[pro_new['State'] == 'KY']

# Display the filtered DataFrame containing providers from KY
print(providers_ky)
```

```python
providers_ky.dtypes
```

```python
providers_ky = pd.DataFrame(providers_ky)

# Function to convert ZIP code to integer (5 digits)
def convert_Zip_Code(Zip_str):
    return int(Zip_str)

# Function to convert phone number to string
def convert_Phone_Number(Phone_Number_str):
    return str(Phone_Number_str)

# Applying the conversion functions to the ZIP and Phone columns in the DataFrame
providers_ky['Zip_Code'] = providers_ky['Zip'].apply(convert_Zip_Code)
providers_ky['Phone_Number'] = providers_ky['Phone'].apply(convert_Phone_Number)

# Display the updated DataFrame with converted ZIP and Phone columns
providers_ky
```
